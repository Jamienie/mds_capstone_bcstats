```python
clean_comments = []
for comment in qual_combined_train['comments']:
    clean_comments.append(preprocessing_text.preprocess_text(comment))
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud,STOPWORDS
plt.figure(figsize=(40,25))

text = clean_comments
cloud_comments = WordCloud(
                          stopwords=STOPWORDS,
                          background_color='black',
                          collocations=False,
                          width=2500,
                          height=1800
                         ).generate(" ".join(text))
plt.axis('off')
plt.title("Words in Comments",fontsize=40)
plt.imshow(cloud_comments)
```
